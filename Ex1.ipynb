````python
from google.colab import drive
import pandas as pd
import plotly.express as px

def connect_drive():
  drive.mount('/content/drive')
def main():
  connect_drive()
  df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/station_Info.csv")
  # with open('/content/drive/MyDrive/Colab Notebooks/station_Info.csv', 'r') as file:
  #     content = file.read()
  df['longitude'] = df['Location'].str.extract('(\d+\.\d+)')[0].astype(float)
  df['latitude'] = df['Location'].str.extract(' (\d+\.\d+)')[0].astype(float)

  fig = px.scatter_mapbox(
    # データフレームおよび緯度・経度の設定
    df,
    lat="latitude",
    lon="longitude",
    hover_name="rowNumber",
    hover_data=['Address'],
    zoom=1,
    center={'lat':34.686567, 'lon':135.52000},
    height=600,
    width=800)
  fig.update_layout(mapbox_style='open-street-map')
  fig.update_layout(title_text="sensor map")
  fig.show()
main()
````
